**Chi-Square** **Test**

```python
import scipy.stats as stats
```

```python
Independent_knn.head()
```

```python
data_table = pd.crosstab(Independent_knn['Calls_To_Customer_Care'],Independent_knn['Dependent'])
print(data_table)
```

```python
Ob_Values = data_table.values
print("Observed_Values:-\n", Ob_Values)
```

```python
val = stats.chi2_contingency(data_table)
print(val)
```

```python
Exp_Val = val[3]
```

```python
from scipy.stats import chi2
chi_square = sum([(o-e)**2./e for o,e in zip(Ob_Values,Exp_Val)])
chi_square_statistic = chi_square[0] + chi_square[1]
```

```python
print('Chi-square statistics:',chi_square_statistic)
```

```python
no_of_rows=len(data_table.iloc[0:5,0])
no_of_columns=len(data_table.iloc[0,0:2])
ddof = (no_of_rows-1)*(no_of_columns-1)
print("The degree of freedom is:",ddof)
alpha =0.05
```

```python
critical_value = chi2.ppf(q=1-alpha,df=ddof)
print("Critical_value is:",critical_value)
```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print("p_value:",p_value)
print("Significance level:",alpha)
print("Degree of freedom:",ddof)
```

If p<=0.05,we reject Null Hypothesis, but here p>0.05,hence we accept the null hypothesis that there is no relation between Calls_To_Customer-Care and Customer_Attrition.Even Chi_square statistics should be > critical value, which is less than critical_value here hence there is no relation between Calls_To_Customer_Care and Customer_Attrition.

Therefore, there is no relation between the Calls_To_Call_Centre and Customer_Attrition variable.
